## Question 1

Use the `open` command to open the file HW03_romeo_juliet.txt. Make sure you've saved this in the same folder as this notebook.  Read the file into a string.

````python
file = open('HW03_romeo_juliet.txt')
````

Now split the string into a list of individual words using the `split()` command.  By default, it splits on whitespace (which includes spaces, tabs, newlines, etc.). Name the list `words`

````python
words = []
for line in file:
    words.extend(line.split())
````

Do the following for your list of words:

1. convert each word to lowercase

    (Hint: Use a loop, or a list comprehension.)
    
        
2. Remove all punctuation (I've included a string of punctuation characters for you called `puncs`)

    (Hint: the method `.replace('!', '')` will remove all exclamation points from a string.  You can loop over the string `puncs` for each word)
    
 
3. Save the new list with the name `words_lower`

````python
puncs = "-.,?!:;'[]"

words_lower = []
for word in words:
    word = word.lower()
    for punc in puncs:
        word = word.replace(punc, '')

    words_lower.append(word.lower())
````

Using `Collections.counter` (like we did in Class 06), count the number of times each word appears

````python
import collections

word_count = collections.Counter(words_lower)
````

What are the 30 most common words?

````python
word_count.most_common(30)
````

````python
word_count['sweet']
````

## Question 2 - Extracting information from a document

Download the file at the url ['http://www.pythonlearn.com/code/mbox-short.txt'](http://www.pythonlearn.com/code/mbox-short.txt) and read the contents into a string named `mail_data` like we did in class.

````python
import requests as re
url = 'http://www.pythonlearn.com/code/mbox-short.txt'
mail_data = re.get(url).text
mail_data
````

Each email header contains a line like the following:

X-DSPAM-Confidence: 0.6961

This has do do with how confident the mail program is about whether the message is spam or not.  

We'd like to extract that floating point number at the end for each message. 

Let's try to extract those numbers without using a regular expression.

Do the following:
1. Split the string `mail_data` into a lines using the `.split()` method
2. Use the `.startswith()` method on each item in the list to identify the lines we want
3. Split the results and keep the number. 
4. Convert that to a float

````python
mail_lines = mail_data.split()
spam_numbers = []
for count, i in enumerate(mail_lines):
    if i.startswith('X-DSPAM-Confidence:'):
        spam_numbers.append(float(mail_lines[count + 1]))
spam_numbers
````

What is the average of the numbers?

````python
from statistics import mean
mean(spam_numbers)
````

Try to extract the same numbers using a regular expression. You could either loop through the list you created, or work on the string before you split it into lines.

````python
import re
spam_nums = re.findall(r'X-DSPAM-Confidence: (.*?)\n', mail_data)
spam_floats = [float(i) for i in spam_nums]
spam_floats
````

## Question 3 - Converting a string of text to a list of sentences

Download the Amazon IPO prospectus text file from the sec website. The address is below.  save the text to a (string) variable named `amazon_ipo`. We did something like this in class.

````python
# download the Amazon IPO text file from the SEC website
import requests
url = 'https://www.sec.gov/Archives/edgar/data/1018724/0000891020-97-000868.txt'

amazon_ipo = requests.get(url).text
print(amazon_ipo)
````

Extract the RISK FACTORS section from the string.

Hint: while you may use a regular expression, you could also split the string a couple of times before and after the section like we did in class.

1. replace all newline characters with spaces using `.replace()`
2. split into sentences using `.split()`

Note: you'll need to make sure you use the right arguments in the `replace` function and `split` function. 

````python
risk_factors_split = amazon_ipo.split('                                  RISK FACTORS')
risk_factors = risk_factors_split[1].split('\n                                       14\n')[0]
risk_factors_no_ws = []
for i in risk_factors.split('.'):
    risk_factors_no_ws.append( i.replace('\n', ' '))
````

Some sentences in the risk factor section are more important than others.  Some include background information, while others outline specific risks. Often 'uncertainty' words like 'might', 'could', or 'may' are included in these types of statements. Make a list of each sentence that includes the word `'may'`. 

Hint: the command `'may' in sentence` will return True if the substring `'may'` is in the string named `sentence`.

You will probably want to use a for loop or a list comprehension as well as a conditional statement

````python
# print(risk_factors.split('.'))
uncertainties = [i for i in risk_factors_no_ws if 'may' in i]
print(uncertainties)
````

How many sentences in the Risk Factor section include the word 'may'?

````python
print(len(uncertainties))
````
